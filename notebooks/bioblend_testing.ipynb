```python
# Fetch available histories
galaxy_instance.histories.get_histories()
history = galaxy_instance.histories.show_history("cbc94c2bcd62ed05")
history
```

```python
# Fetch two dataset collections I've created
dsc1 = galaxy_instance.histories.show_matching_datasets(history["id"], name_filter="Collection 1")[0]
dsc2 = galaxy_instance.histories.show_matching_datasets(history["id"], name_filter="Collection 2")[0]
```

```python
# Generate a new DataSet collection that is actually a list:list
collection_description = {
    'collection_type': 'list:list',
    'element_identifiers': [
        {'id': dsc1["id"], 'name': 'Collection 1', 'src': 'hdca'},
        {'id': dsc2["id"], 'name': 'Collection 2','src': 'hdca'}],
             'name': 'Collection 3 (list:list)'}
dsc_list_list = galaxy_instance.histories.create_dataset_collection(history_id=history["id"], collection_description=collection_description)
```
